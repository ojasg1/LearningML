# Importing Numpy,Pandas,Matplotlib and Seaborn Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading Iris Dataset

```python
df = pd.read_csv("iris.csv")
df.head()
```

# Displaying first 5 entries

```python
df.head()
```

# Displaying the info of df

```python
df.info()
```

# Grouping the entries by species and adding an aggregate function max to print only maximum value of species.

```python
byspecies = df.groupby(by='species').max()
byspecies.head()
```

```python
sns.scatterplot(data=df, x=df['sepal_length'], y=df['sepal_width'], hue=df['species'])
```

Setosa species has lower sepal length but higher sepal width.
Sepal length of setosa species ranged between 4.0 to 6.0 and sepal width between 2.5 and 4.5.
Sepal length of versicolor species ranged between 5.0 to 7.0 and sepal width between 2.0 to 3.5.
Sepal length of virginica species ranged between 5.0 and 8.0 and sepal width between 2.0 and 4.0.

```python
sns.countplot(x = 'species', data = df)
```

Here we can see the dataset is a balanced dataset as it contains same number of entries for all three species.

```python
plt.figure(figsize=(15,5))
sns.countplot(x = 'sepal_length', data = df, hue = 'species')
plt.legend(bbox_to_anchor=(1,1))
```

Here we can observe maximum sepal length of setosa is 5.8 and maximum number of setosa are lengthened 5.0 and 5.1.
Maximum sepal length of versicolor is 7.0 and maximum number of versicolor are lengthened between 5.5 and 5.7.
Maximum sepal length of virginica is 7.9 and maximum number of virginica lengthened 6.3.
Setosa lengthened the lowest and virginica lengthened the highest in comparison.

```python
plt.figure(figsize=(15,5))
sns.countplot(x = 'petal_length', data = df, hue = 'species')
plt.legend(bbox_to_anchor=(1,1))
```

Here, Setosa has the lowest petal length and virginica has the highest petal length. Maximum of Setosa are of 1.5 petal length,maximum number of versicolor are of 4.5 petal length and maximum number of virginica are of 5.1 petal length. 

```python
plt.figure(figsize=(15,5))
sns.countplot(x = 'sepal_width', data = df, hue = 'species')
plt.legend(bbox_to_anchor=(1,1))
```

Here, Versicolor is the lowest and setosa is the highest in sepal width comparison.

```python
plt.figure(figsize=(15,5))
sns.countplot(x = 'petal_width', data = df, hue = 'species')
plt.legend(bbox_to_anchor=(1,1))
```

Here, Setosa has the lowest petal width and virginica has the highest petal width.

```python
plt.figure(figsize=(10,5))
sns.lmplot(x = 'petal_length' , y = 'sepal_length' , data=df, hue='species')
```

Here, the above graph shows the range of sepal length and petal length according to species.

```python
byspecies['petal_length'].plot()
```

Here, Petal length increases in a linear manner.

```python
df['species'].value_counts().plot.pie()
```

The above pie chart confirms the conclusion that the iris datset is a balanced dataset.

```python
byspecies['petal_length'].plot.pie()
```

Here, we can see Virginica has the highest petal length and setosa has the lowest.

```python
sns.boxplot(x="species", y="petal_length", data=df)
```

Here, Setosa has the lowest petal length and virginica has the highest petal length.

```python
sns.boxplot(x="species", y="sepal_length", data=df)
```

Here,Setosa has the lowest Sepal length and virginica has highest sepal length.

```python
sns.boxplot(x="species", y="sepal_width", data=df)
```

Here, setosa has the highest sepal width and versicolor has the lowest sepal width.

```python
sns.boxplot(x="species", y="petal_width", data=df)
```

Here we can observe setosa has the lowest petal width and virginica has the highest petal width.

```python
sns.set_style('darkgrid')
sns.pairplot(data = df, hue = "species")
```

Here we can easily distinguish Setosa from versicolor and virginica. Versicolor and Virginica overlap each other. So, we can say versicolor and virginica are a little similar to each other. 

```python
plt.figure(figsize=(15,7))
sns.clustermap(byspecies,annot=True)
```

Here, we can observe that sepal length and petal length are more in comparison to sepal width and petal width except for setosa whose petal length is less. This shows versicolor and virginica species are bigger in length and smaller in width.
